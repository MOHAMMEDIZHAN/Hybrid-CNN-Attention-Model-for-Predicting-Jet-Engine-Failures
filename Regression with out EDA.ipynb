# REGRESSION MODEL
The goal of this project was to predict the remaining useful life of turbo fan engines, previously we converted it into a classification task, where the accuray of the model was failrly good. But here we'll treat this as a regression problem. The goal will be to train a ML model without any EDA or hyperparameter tunings and see how accurate does it predict the RUL. If the predictions will turn out to be faulty or unreliable, we'll move on to do the Exploratory Data analysis, feature selection part and hyperparameter tuning.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('RUL_Train_001.csv')
```

```python
df
```

```python
df[df['RUL']==0] # Gives data on how many engine's Run-to-failure data is available
```

## Train Test Split

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle=False)
```

## Random Forest Regression

```python
# Initializing the Model with arbitary parameters
regressor = RandomForestRegressor()

# Training the model
regressor.fit(X_train,y_train)
```

## Training Plot

```python
# Using the trained model to check the Training data
y_pred = regressor.predict(X_train)

print('RMSE on Train set:', np.sqrt(metrics.mean_squared_error(y_train ,y_pred)))
print('MAE on Train set:', (metrics.median_absolute_error(y_train ,y_pred)))
print('R_2 on Train set:', (metrics.r2_score(y_train ,y_pred)))

# Plot the train data
plt.scatter(X_train['Cycle'][0:192],y_pred[0:192])
plt.scatter(X_train['Cycle'][0:192],np.array(y_train)[0:192])
plt.ylabel('RUL')
plt.xlabel('CYCLE')
plt.legend(['Prediction', 'Actual'], loc='upper right')
plt.show()


```

## Test Plot

```python
# Using the trained model to check the Test data
y_pred = regressor.predict(X_test)

print('RMSE on Test set:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('MAE on Train set:', (metrics.median_absolute_error(y_test, y_pred)))
print('R_2 on Train set:', (metrics.r2_score(y_test, y_pred)))

# Plotting the test data
plt.scatter(X_test['Cycle'][100:200],y_pred[100:200])
plt.scatter(X_test['Cycle'][100:200],np.array(y_test)[100:200])
plt.legend(['Prediction', 'Actual'], loc='upper right')
plt.ylabel('RUL')
plt.xlabel('CYCLES')
plt.show()
```
